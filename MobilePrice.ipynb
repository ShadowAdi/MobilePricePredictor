```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
mobile_df=mobile_df.sample(frac=1)
```

```python
mobile_df.info()
```

```python
X=mobile_df.drop(columns=["Price in Rupees"])
y=mobile_df["Price in Rupees"]
```

```python
X.columns
```

```python
cat_cols=[]
for i in X.columns:
    if X[i].dtype=="O":
        cat_cols.append(i)
```

```python
num_cols=[]
for i in X.columns:
    if X[i].dtype!="O":
        if (len(X[i].unique())<3):
            cat_cols.append(i)
        else:
            num_cols.append(i)
```

```python
cat_cols
```

```python
num_cols
```

```python
preprocessor=ColumnTransformer(
    transformers=[
        ("num",StandardScaler(),num_cols),
        ("cat",OneHotEncoder(),cat_cols)
    ]
)
```

```python
pipeline=Pipeline(steps=[("preprocessor",preprocessor)])
```

```python
X["Model"].unique()
```

```python
X
```

```python
X
```

```python
X_transformed=pipeline.fit_transform(X),
```

```python
X_transformed
```

```python
df_transformed = pd.DataFrame(pipeline.fit_transform(X),
                               columns=num_cols + list(preprocessor.named_transformers_['cat'].get_feature_names_out(cat_cols)))  
df_transformed
```

```python
X_train,X_test,y_train,y_test=train_test_split(df_transformed,np.log1p(y),test_size=0.23,random_state=32)
```

```python
y_train.shape
```

```python
X_test
```

```python
from sklearn.ensemble import RandomForestRegressor,BaggingRegressor,AdaBoostRegressor,GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from xgboost import XGBRegressor
from sklearn.ensemble import StackingRegressor

```

```python
y_train
```

```python
def model_trainer(model):
    np.random.seed(32)
    modelfitted=model().fit(X_train,y_train)
    ans_pred=modelfitted.predict(X_test)
    return {
        "r2_score":r2_score(y_test,ans_pred),
        "mae_score":mean_absolute_error(y_test,ans_pred),
        "mse_score":mean_squared_error(y_test,ans_pred),
        "modelname":model()
    }
```
